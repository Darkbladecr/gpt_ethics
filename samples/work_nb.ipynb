```python
import pandas as pd

df = pd.read_csv("SJT responses from ChatGPT.csv")
```

```python
df.columns
```

```python
letter_based_questions = df[df['type']=='Ranking']
letter_based_questions.head()
```

```python
def score_response(true_answer: str, model_prediction: str):
    assert(isinstance(model_prediction, str))
    assert(isinstance(true_answer, str))
    score = 0
    for predicted_index in range(0, len(model_prediction)):
        # index difference between the two strings of the same letter
        # determines the score
        letter = model_prediction[predicted_index]
        # print(letter)
        true_index = true_answer.find(letter)
        # print(true_index, predicted_index)
        score += 4 - abs(true_index-predicted_index)
    return score
        
        
score_response("ABCDE", "BACDE")
```

```python
# letter_based_questions['score'] = score_response.apply
letter_based_questions['gpt3.5_scored'] = letter_based_questions.apply(lambda row: score_response(row['answer'], row['gpt3.5']), axis=1)

letter_based_questions
```
